Otro tipo de grfico comnmente utilizado es el grfico de dispersin, un pariente cercano del grfico de lneas, y que ya conoces de nuestros anlisis bivariantes del sprint anterior.

En lugar de que los puntos estn unidos por segmentos de lnea, aqu los puntos se representan individualmente con un punto, crculo u otra forma.

````python
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

### Grficos de dispersin con ``plt.plot`` y ``ax.plot``

En la sesin anterior, vimos ``plt.plot``/``ax.plot`` para producir grficos de lneas.
Resulta que esta misma funcin tambin puede producir grficos de dispersin, con slo cambiar algunos argumentos:
```

````python
x = np.linspace(0,10,20)
y = np.sin(x)
fig,ax = plt.subplots(1,2)
ax[0].plot(x,y,"-")
ax[1].plot(x,y,"o")
````

El tercer argumento en la llamada a la funcin es un carcter que representa el tipo de smbolo utilizado para el trazado. Al igual que puedes especificar opciones como ``'-'``, ``'--'`` para controlar el estilo de lnea, el estilo del marcador tiene su propio conjunto de cdigos de cadena cortos.  

La lista completa de smbolos disponibles se puede ver en la [documentacin de ``plt.plot``](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html), o en la documentacin en lnea de Matplotlib. La mayora de las posibilidades son bastante intuitivas, y mostraremos algunos de los ms comunes aqu:

````python
markers = ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']
rng = np.random.RandomState(0)
for marker in markers:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label=f"marker='{marker}'")
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
````

Para an ms posibilidades, estos cdigos de caracteres pueden ser utilizados junto con cdigos de lnea y color para trazar puntos junto con una lnea que los conecte:

````python
plt.plot(x,y,"-ok")
````

### Grficos de dispersin con ``plt.scatter`` y ``ax.scatter``

Un segundo mtodo ms potente para crear grficos de dispersin es la funcin ``plt.scatter``, que se puede usar de manera muy similar a la funcin ``plt.plot`` y que es el que se usa en las funciones que empleamos en los anlisis:


````python
plt.scatter(x,y);
````


**La diferencia principal de ``plt.scatter`` con ``plt.plot`` es que se puede utilizar para crear grficos de dispersin donde las propiedades de cada punto individual (tamao, color de relleno, color de borde, etc.) pueden ser controladas individualmente o asignadas a datos.** De hecho, gracias a esto creamos los diagramas de burbujas donde el tamao de los puntos depende de una tercera variable.

Vemoslo creando un grfico de dispersin aleatorio con puntos de muchos colores y tamaos.
Para poder ver mejor los resultados superpuestos, tambin usaremos la palabra clave ``alpha`` para ajustar el nivel de transparencia:


````python
rng = np.random.RandomState(0)
plt.grid(False)
x = rng.randn(100) # edades
y = rng.randn(100) # alturas
colors = rng.rand(100) # pais
sizes = 1000 * rng.rand(100) # numero personas

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis');
plt.colorbar();
````

Para verlo con un dataset real, usemos otro de los dataset "hello, world" junto con el del titanic, el iris dataset, en el que vamos a mostrar las grficas de tres tipos de flores de la misma familia enfrentando la longitud del ptalo y del spalo y pintando cada tipo de flor de un color, y regulando el tamao de los puntos con el ancho del spalo:

````python
df_iris = pd.read_csv("./data/iris.csv")
df_iris
````

````python
plt.scatter(df_iris["sepal length (cm)"], df_iris["petal length (cm)"],c = df_iris["target"],\
            alpha=0.5,s=df_iris["petal width (cm)"]*100)
plt.xlabel("sepal length (cm)")
plt.ylabel("petal length (cm)")
plt.title("Iris Dataset")
plt.colorbar()
````

Podemos ver que este grfico de dispersin nos ha dado la capacidad de **explorar simultneamente cuatro dimensiones diferentes de los datos:**
la ubicacin (x, y) de cada punto corresponde a la longitud y anchura del spalo, el tamao del punto est relacionado con la anchura del ptalo, y el color est relacionado con la especie particular de flor.
Grficos de dispersin multicolores y multifuncionales como este pueden ser tiles tanto para la exploracin como para la presentacin de datos.
