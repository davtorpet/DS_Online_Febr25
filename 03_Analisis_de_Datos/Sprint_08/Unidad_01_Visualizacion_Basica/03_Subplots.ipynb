# Visualizacin con Matplotlib: Subplots

Como ya sabemos de nuestros anlisis del sprint anterior,  **muchas veces es til comparar diferentes vistas de datos lado a lado.**
Para este fin, Matplotlib tiene el concepto de *subplots*: grupos de ejes ms pequeos que pueden existir juntos dentro de una sola figura.
Estos subplots podran ser inserciones, cuadrculas de grficos u otros diseos ms complicados y adems no tienen por qu ser del mismo tipo.  

En esta seccin exploraremos como crear subplots en Matplotlib.
```

Importamos como siempre:

````python
# %matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")
import numpy as np

x = np.linspace(0, 10, 100)
````


### "A mano"

El mtodo ms bsico para crear un eje es usar la funcin ``plt.axes``.
Como hemos visto anteriormente, de forma predeterminada esto crea un objeto de eje estndar que llena toda la figura.
``plt.axes`` tambin toma un argumento opcional que es una lista de cuatro nmeros en el sistema de coordenadas de la figura.
Estos nmeros representan ``[izquierda, inferior, ancho, altura]`` en el sistema de coordenadas de la figura, que vara de 0 en la esquina inferior izquierda de la figura a 1 en la esquina superior derecha de la figura.

Por ejemplo, podramos crear un eje secundario en la esquina superior derecha de otro eje estableciendo la posicin *x* y *y* en 0.65 (es decir, comenzando en el 65% del ancho y el 65% de la altura de la figura) y las extensiones *x* y *y* en 0.2 (es decir, el tamao del eje es el 20% del ancho y el 20% de la altura de la figura):


````python
ax1 = plt.axes()
ax2 = plt.axes([0.1, 0.1, 0.2, 0.3])
ax1.plot(x,2*x)
ax2.plot(x,np.sin(x))
````

El equivalente de este comando dentro de la interfaz orientada a objetos es ``fig.add_axes()``. Usemos esto para crear dos ejes apilados verticalmente:

````python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 1, 1, 2],
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 1.5, 1],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
````


### Con cuadrculas sencillas

Las columnas o filas alineadas de *subplots* o grficas formando cuadrculas o matrices son una necesidad lo suficientemente comn como para que Matplotlib tenga varias rutinas de conveniencia que las hacen fciles de crear.

El nivel ms bajo de stas es ``plt.subplot()``, este comando tiene tres argumentos enteros: el nmero de filas, el nmero de columnas y el ndice de la grfica que se crear en este esquema, que va desde la esquina superior izquierda hasta la esquina inferior derecha.


Vamos a crear una figura con una matriz de subplots o grficas de 3x5, y las numeraremos con su ndice para que te quede ms claro:

````python
plt.figure(figsize=(15,10))
for i in range(1,16):
    plt.subplot(3,5,i) #Empieza a contar en 1 ,i 
    plt.text(0.5,0.5,str((3,5,i)),fontsize=16, ha='center')
````

El comando ``plt.subplots_adjust`` se puede usar para ajustar el espaciado entre estos grficos.
El siguiente cdigo utiliza el comando orientado a objetos equivalente, ``fig.add_subplot()``:

````python
plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.4, wspace=0.4) #Espacio entre subplots
for i in range(1,16):
    plt.subplot(3,5,i) #Empieza a contar en 1 ,i 
    plt.text(0.5,0.5,str((3,5,i)),fontsize=16, ha='center')
````

Hemos utilizado los argumentos ``hspace`` y ``wspace`` de ``plt.subplots_adjust``, que especifican el **espaciado a lo largo de la altura y anchura de la figura**, en unidades del tamao del subtrazado (en este caso, el espacio es el 40% de la anchura y altura del subtrazado).


### La cuadrcula completa de una vez

**La aproximacin anterior puede volverse bastante tediosa si queremos crear una cuadrcula o matriz grande, especialmente si se desea ocultar las etiquetas del eje x y y en las tramas internas.**  

Para este propsito, ``plt.subplots()`` es la herramienta ms fcil de usar (nota la ``s`` al final de ``subplots``). En lugar de crear una sola subtrama, esta funcin crea una cuadrcula completa de subtramas en una sola lnea, devolvindolas en un array de NumPy.
Los argumentos son el nmero de filas y el nmero de columnas, junto con palabras clave opcionales ``sharex`` y ``sharey``, que le permiten especificar las relaciones entre diferentes ejes.  


Aqu crearemos una cuadrcula de $2 \times 3$ subtramas, donde todos los ejes en la misma fila comparten su escala en el eje y, y todos los ejes en la misma columna comparten su escala en el eje x:
```

````python
fig,ax = plt.subplots(2,3,sharex="col",sharey="row") #Subplots en una figura
````

````python
# Los axes estn en un array bidimensional accedido por [fila,columna] tipo python, empezando en cero
for i in range(2):
    for j in range(3):
        ax[i,j].text(0.5,0.5,str((i,j)),fontsize=18, ha='center')
fig
````

Si queremos acceder a cualquiera de los grficos podemos acceder a partir del elemento del array de objetos en "ax" (tal como hemos aadido los textos):

````python
ax[1,1].plot(x,np.sin(x));
fig
````

````python
# Te dejo aqu como haramos para quitar el texto:
fig, ax = plt.subplots(2, 3, sharex = "col", sharey = "row")
textos = np.full(fill_value= None, shape= ax.shape)
for i in range(2):
    for j in range(3):  
        textos[i,j] = ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
textos[0,2].remove()
textos[1,1].remove()
ax[1,1].plot(x,np.sin(x));
````
